##5. Set Up Great Expectations Context and Data Source

```python
# Write code here
context = gx.get_context()  # Initialize Great Expectations context
data_source = context.data_sources.add_pandas("pandas")  # Add a pandas data source
data_asset = data_source.add_dataframe_asset(name="pd_dataframe_asset")  # Define the data asset
```

##6. Define and Create a Data Batch

```python
# Write code here
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch_definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})  # Retrieve the batch of data
```
